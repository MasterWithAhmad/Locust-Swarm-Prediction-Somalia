```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_style("whitegrid")

# Define numerical columns
numerical_cols = ['PPT', 'TMAX', 'SOILMOISTURE']

# Plot distributions
plt.figure(figsize=(15, 5))
for i, col in enumerate(numerical_cols):
    plt.subplot(1, 3, i + 1)
    sns.histplot(df[col], bins=30, kde=True, color='skyblue')
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

## **Insights** 

This visualization helps us understand how precipitation, temperature, and soil moisture are distributed.
Skewness or outliers in these distributions may indicate data issue

We can also check for normality, which is important for certain models.

```python
# Filter Somalia data
df['COUNTRYNAME']=df['COUNTRYNAME'].str.strip()
somalia_df = df[df['COUNTRYNAME'] == 'Somaliland']

# Plot distributions for Somalia
plt.figure(figsize=(15, 5))
for i, col in enumerate(numerical_cols):
    plt.subplot(1, 3, i + 1)
    sns.histplot(somalia_df[col], bins=30, kde=True, color='salmon')
    plt.title(f'Somalia: Distribution of {col}')
plt.tight_layout()
plt.show()
```

```python
# Encode categorical target variable
df['LOCUSTPRESENT'] = df['LOCUSTPRESENT'].map({'no': 0, 'yes': 1})

# Compute correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(df[numerical_cols + ['LOCUSTPRESENT']].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Matrix")
plt.show()
```

## *Insights*
After analyzing the dataset, we observed that temperature and soil moisture exhibit moderate correlation with locust presence, while precipitation has a weaker correlation.
The correlation matrix highlights that no single variable is an absolute determinant of locust presence, emphasizing the need for multi-variable predictive modeling.

```python
# Countplot for Somalia regions
plt.figure(figsize=(12, 5))
sns.countplot(data=somalia_df, x='REGION', hue='LOCUSTPRESENT', palette='coolwarm', order=somalia_df['REGION'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Locust Presence Across Somaliaâ€™s Regions')
plt.show()
```

## **Insights** 
This visualization highlights which regions in Somalia experience more locust outbreaks.

Could be useful for targeted interventions and risk assessments.

# ### *3.3 Handling Outliers*

```python
# Using the IQR method to identify outliers
Q1 = df[numerical_cols].quantile(0.25)
Q3 = df[numerical_cols].quantile(0.75)
IQR = Q3 - Q1

# Defining lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtering out outliers
df = df[~((df[numerical_cols] < lower_bound) | (df[numerical_cols] > upper_bound)).any(axis=1)]
```
