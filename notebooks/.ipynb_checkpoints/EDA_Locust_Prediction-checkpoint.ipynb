# **Exploratory Data Analysis (EDA) for Locust Prediction Dataset**
    
This notebook performs a comprehensive **Exploratory Data Analysis (EDA)** on the Locust Prediction Dataset.  
The analysis includes data understanding, exploration, preprocessing, feature engineering, model training, and evaluation.


```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from imblearn.over_sampling import SMOTE  # For handling class imbalance
import warnings
warnings.filterwarnings('ignore')

# Set Seaborn style
sns.set_style("whitegrid")
```

```python
# 1. Data Understanding: Load and Describe Dataset

# Load dataset
file_path = "../data/locust_dataset.csv"
df = pd.read_csv(file_path)

# Display first few rows
df.head()
```

```python
# 1.1 Summary Statistics
df.info()  # Dataset structure

# Display basic statistics
df.describe()
```

```python
# 2. Data Exploration: Visualizing Feature Distributions

# Distribution of Numerical Features
df.hist(figsize=(12, 8), bins=30, edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()
```

```python
# 2.2 Correlation Analysis

plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Matrix")
plt.show()
```

```python
# 3. Data Preprocessing

# Check for missing values
print("Missing values per column:")
print(df.isnull().sum())

# Handle missing values (if any)
df = df.dropna()

# Remove duplicate rows
df = df.drop_duplicates()

print(f"Dataset shape after preprocessing: {df.shape}")
```

```python
# 4. Feature Engineering

# Creating Season Category
df['SEASON'] = df['STARTMONTH'].apply(lambda x: 'Dry' if x in [1, 2, 11, 12] else 'Wet')

# Categorizing Temperature Levels
df['TEMP_CATEGORY'] = pd.cut(df['TMAX'], bins=[0, 20, 30, 50], labels=['Cold', 'Moderate', 'Hot'])

# Categorizing Precipitation Levels
df['PPT_CATEGORY'] = pd.cut(df['PPT'], bins=[-1, 5, 20, 100], labels=['Low', 'Medium', 'High'])

# Display updated dataset
df.head()
```

```python
# 5. Handling Class Imbalance

# Check target class distribution
print(df['LOCUSTPRESENT'].value_counts())

# Convert categorical target to numerical
label_encoder = LabelEncoder()
df['LOCUSTPRESENT'] = label_encoder.fit_transform(df['LOCUSTPRESENT'])

# Define features and target
X = df.drop(columns=['LOCUSTPRESENT', 'COUNTRYNAME', 'REGION'])  # Drop categorical columns for simplicity
y = df['LOCUSTPRESENT']

# Apply SMOTE to balance the classes
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

print("Class distribution after SMOTE:", np.bincount(y_resampled))
```

```python
# 6. Model Training

# Splitting Data
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Train Random Forest Classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

print("Random Forest Model Trained Successfully")
```

```python
# 7. Model Evaluation

# Predictions
y_pred = rf_model.predict(X_test)

# Evaluation Metrics
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt="d", cmap="Blues", xticklabels=["No Locust", "Locust"], yticklabels=["No Locust", "Locust"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
# 8. Feature Importance Analysis

# Extract feature importances
importances = rf_model.feature_importances_
feature_names = X.columns

# Create DataFrame
importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances}).sort_values(by="Importance", ascending=False)

# Plot Feature Importance
plt.figure(figsize=(10, 5))
sns.barplot(x=importance_df["Importance"], y=importance_df["Feature"], palette="viridis")
plt.title("Feature Importance in Random Forest Model")
plt.show()
```
